```python
import itertools
import json
```

```python
queen_bee_concentrations = [0.15]
worker_bee_concentrations = [0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5]
diffusion_coefficients = [0.05, 0.15, 0.25, 0.35, 0.45, 0.55]
worker_bee_thresholds = [0.005, 0.5]

for i, combination in enumerate(itertools.product(queen_bee_concentrations, worker_bee_concentrations, diffusion_coefficients, worker_bee_thresholds)):
    print(i, combination)
```
