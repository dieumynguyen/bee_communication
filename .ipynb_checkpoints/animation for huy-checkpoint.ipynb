```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML
import pandas as pd
```

```python
file = "Waggle Dance Video Waggle log.xlsx"
```

```python
data = pd.read_excel(file)
position = data["Unnamed: 4"]
intensity = data["Unnamed: 6"]
```

```python
# Convert to numeric xy
ref = 'abcdefg'
x_pos = []; y_pos = []

for pos in position[1:]:
    y = int(pos[1].lower())
    y_pos.append(y)
    
    # Get index from ref string as numeric x value
    x = pos[0].lower()
    new_x = ref.index(x)
    x_pos.append(new_x)

# List comprehension way, saving in array of tuples
# xy_data = np.array([(ref.index(pos[0].lower()), int(pos[1].lower())) for pos in position[1:]])
```

```python
# Make animation - basically a bunch of images for each
# timestep, then make a gif on bash/terminal
my_dpi=96
for i in range(len(x_pos)):
    fig = plt.figure(figsize=(2,2), dpi=my_dpi)
    plt.scatter(x_pos[i], y_pos[i], s=intensity[i+1]*100, 
                c="orange", alpha=0.6, edgecolors="orange", 
                linewidth=2)
    
    # Add titles (main and on axis)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("t: " + str(i) )
    plt.ylim(1, 6)
    plt.xlim(0, 6)
    
    # Save image of each timestep
    filename='huy_bee/bee_position_t{0:03}.png'.format(i)
    plt.savefig(filename, dpi=96)
```
